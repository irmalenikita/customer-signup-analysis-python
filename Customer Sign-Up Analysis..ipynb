# Customer Sign-Up Analysis

Objective: Audit and analyse the customer-signups.csv dataset to identify data quality issues and derive business insights into user acquisition, plan selection, and marketing opt-in patterns

- Import Libraries and Load Dataset

```python
import pandas as pd
import numpy as np

df=pd.read_csv(r"C:\Users\mymac\Downloads\customer_signups.csv")
df
```

```python
df.head()
```

Identify Missing Values, Duplicates, and Data Types

```python
# Check data types
print(df.dtypes)
```

```python
#Count missing values
print(df.isnull().sum())
```

```python
#Count duplicates by customer_id
duplicates = df.duplicated(subset='customer_id').sum()
print(f'Duplicates: {duplicates}')
```

```python
df['signup_date'] = pd.to_datetime(df['signup_date'], format='%d-%m-%y', errors='coerce')
df['signup_date'].head()
df['signup_date'].dtypes
df
```

```python
#Remove rows with missing or invalid customer_id and eliminate duplicates.

df = df[df['customer_id'].notna() & (df['customer_id'].str.strip() != '')]
df
```

```python
#Correct inconsistent plan names and unify text case for reliable analysis.
#PLAN_SELECTED

df.loc[:, 'plan_selected'] = df['plan_selected'].astype(str).str.strip().str.lower()
df.loc[:, 'plan_selected'] = df['plan_selected'].replace(['','unknownplan', ' ', 'none', 'nan'],'Unknown')
df.loc[:, 'plan_selected'] = df['plan_selected'].replace({
    'basic': 'Basic',
    'pro': 'Pro',
    'premium': 'Premium',
    'prem': 'Premium'
})

df['plan_selected'].value_counts()
```

```python
#Fix inconsistent entries (e.g., “M”, “Females”, “Nonbinary”) and replace blanks with “Unknown”.
#Gender
df.loc[:,'gender'] = df['gender'].astype(str).str.strip().str.title()


df.loc[df['gender'].str.isnumeric(), 'gender'] = 'Unknown' ##Replace invalid numeric values with 'Unknown'

#Fix inconsistent variations
df.loc[:,'gender'] = df['gender'].replace({
    'Females': 'Female',
    'Femail': 'Female',
    'F': 'Female',
    'M': 'Male',
    'Males': 'Male',
    'Non Binary': 'Non-Binary',
    'Nonbinary': 'Non-Binary',
    'Nb': 'Non-Binary',
    'Others': 'Other'
})

df.loc[:,'gender'] = df['gender'].replace(['', ' ', 'None', 'Nan', 'NaN'], 'Unknown') #Replace blanks or missing with 'Unknown'

df['gender'].value_counts() ##Check cleaned values
```

```python
#Validate and Clean email Field
#email column:
#Strip spaces and replace blanks or None
df.loc[:,'email'] = df['email'].astype(str).str.strip()
df.loc[:,'email'] = df['email'].replace(['', ' ', 'None', 'nan', 'NaN'], np.nan)

#Validate email format using simple regex
import re
def validate_email(email):
    if pd.isna(email):
        return 'Unknown'
    pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'
    if re.match(pattern, email):
        return email
    else:
        return 'Unknown'

df.loc[:,'email'] = df['email'].apply(validate_email)

# Check results
df['email'].value_counts()
```

* Missing emails, regions, ages, names

* Duplicates (if any)

* Columns that need type conversion (like signup_date)

```python
#Fix non-numeric or invalid age entries, convert to numeric type, and replace missing values with median age.
#Clean up Age column

df['age'] = pd.to_numeric(df['age'], errors='coerce')
df.loc[(df['age'] < 10) | (df['age'] > 100), 'age'] = np.nan
median_age = df['age'].median()
df.loc[:,'age'] = df['age'].fillna(median_age)             ##Fill missing ages with median
print(df['age'].dtype)
```

```python
#Region
df.loc[:,'region'] = df['region'].astype(str).str.strip()
df.loc[:,'region'] = df['region'].replace(['', ' ', 'nan', 'Na', 'None', 'N/A'],'Unknown')   #fill wiht Unknown

df['region'].value_counts()               #Check results
```

```python
#Name 
df.loc[:,'name'] = df['name'].astype(str).str.strip()
df.loc[:,'name'] = df['name'].replace(['', ' ', 'None', 'nan', 'NaN'],'Unknown')

df['name'].value_counts()   #Check results
```

```python
#Marketing_opt

df.loc[:,'marketing_opt_in'] = df['marketing_opt_in'].astype(str).str.strip()

df.loc[:,'marketing_opt_in'] = df['marketing_opt_in'].replace(['', ' ','Nil','NIL','None', 'none', 'nan', 'NaN'], 'Unknown')

df['marketing_opt_in'].value_counts() #check the unique values
```

```python
## Source

df.loc[:,'source'] = df['source'].astype(str).str.strip().str.title()
df.loc[:,'source'] = df['source'].replace(['??', 'nan','Nan', 'Na', 'N/A', '', 'None'],'Unknown')  #Fill with 'Unknown'


df['source'].value_counts()             #Check values
```

# Create Final Cleaned Dataset

Remove all rows containing any “Unknown” placeholders to create a high-quality dataset.

```python
# Remove rows where any column has the string 'Unknown'
df_cleaned = df[~df.isin(['Unknown']).any(axis=1)]

# Check remaining rows
print(df_cleaned.shape)
```

# Business Questions

```python
#Q.1 Which acquisition source brought in the most users last month?

# Find last month in the data
last_month = df['signup_date'].max().month

# Filter data for that month
last_month_data = df[df['signup_date'].dt.month == last_month]

# Count sign-ups by source
print(last_month_data['source'].value_counts())
print("Last month in dataset:", df['signup_date'].max().strftime('%B %Y'))

```

```python
#Q.2 Are older users more or less likely to opt in to marketing?
#Average opt-in rate by age

print(df.groupby('marketing_opt_in')['age'].mean())

```

```python
#Q.3 Which plan is most commonly selected, and by which age group?
print(df['plan_selected'].value_counts())
print(df.groupby('plan_selected')['age'].mean())
```
